```python
#정의 부분
result = []
r_P = [n for n in range(1,14)]
r_K = [n for n in range(1,14)]
r_H = [n for n in range(1,14)]
r_T = [n for n in range(1,14)]

#함수 부분
def Div(x) :
  for i in range(0,len(x),3) :
    result.append(x[i])

  for i in range(len(x)) :
    if x[i] in result :
      if x[i] == "P" :
        Int = int(x[i+1]+x[i+2])
        if Int not in r_P :
          print("GRESKA")
          return
        else :
          r_P.remove(Int)

      elif x[i] == "K" :
        Int = int(x[i+1]+x[i+2])
        if Int not in r_K :
          print("GRESKA")
          return
        else :
          r_K.remove(Int)

      elif x[i] == "H" :
        Int = int(x[i+1]+x[i+2])
        if Int not in r_H :
          print("GRESKA")
          return
        else :
          r_H.remove(Int)

      elif x[i] == "T" :
        Int = int(x[i+1]+x[i+2])
        if Int not in r_T :
          print("GRESKA")
          return
        else :
          r_T.remove(Int)

  print(len(r_P), len(r_K), len(r_H), len(r_T))

#메인 부분
if __name__ in "__main__" :
  x = input()
  Div(x)
```

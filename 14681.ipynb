```python
def four(x,y) :
  if (x > 0) :
    if (y > 0) :
      print(1)
      return
    else :
      print(4)
      return
  else :
    if (y > 0) :
      print(2)
      return
    else :
      print(3)
      return

if __name__ == "__main__" :
  x = int(input())
  y = int(input())
  four(x,y)
```

## 8958번: OX퀴즈

```python
ask1 = input()
num = int(ask1)

score = []

for i in range(num) :
    ask2 = input()
    score.append(str(ask2))

# 확인용
# print(score)

cnt = 0
result = []

for i in range(len(score)) :
    for j in range(len(score[i])) :
        if score[i][j] == 'O' :
            cnt += 1
            result.append(cnt)
        else :
            cnt = 0
            result.append(cnt)

    print(sum(result))
    cnt = 0
    result.clear()
```

## 1978번: 소수 찾기

```python
T = int(input())
Q = input()
```

```python
inp = list(map(int, Q.split()))

result = 0
for i in range(T) :
  if inp[i] == 1 :
    continue
  p = True
  for j in range(2, inp[i]) :
    if inp[i] % j == 0 :
      p = False
      break
  if p :
    result += 1

print(result)
```

#### 1978번 코드는 gpt의 도움을 받아서 풀었다. for문의 범위가 헷갈렸고, 백준에서 요구하는 "나누어 떨어지면 탈락" 이라는 부분을 알아내지 못해 처음엔 append를 하고 마지막에 소수인지 아닌지를 판별했다면, 후에는 지피티의 도움을 받아 초기에 True로 설정하고 조건을 만족하지 않으면 False로 바꿔 합산하는 코드로 수정했다.
